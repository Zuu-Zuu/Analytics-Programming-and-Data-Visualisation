```python
covid_data = pd.read_csv("C:\\Users\\Admin\\WHO_COVID19_Data\\WHO-COVID-19-global-daily-data.csv")
covid_data.head()
covid_data.isna().sum()
```

```python

covid_data['Date_reported'] = pd.to_datetime(covid_data['Date_reported'])

covid_data['New_cases'] = covid_data['New_cases'].fillna(0)
covid_data['New_deaths'] = covid_data['New_deaths'].fillna(0)

global_trends = covid_data.groupby('Date_reported')[['New_cases', 'New_deaths']].sum().reset_index()


plt.figure(figsize=(14, 6))
plt.plot(global_trends['Date_reported'], global_trends['New_cases'], label='New Cases', alpha=0.8)
plt.plot(global_trends['Date_reported'], global_trends['New_deaths'], label='New Deaths', alpha=0.8, color='red')

plt.title('Global COVID-19 Trends: New Cases and Deaths', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
vaccination_data = pd.read_csv("C:\\Users\\Admin\\WHO_COVID19_Data\\vaccination-data.csv")
vaccination_data['WHO_REGION'] = vaccination_data['WHO_REGION'].fillna('Unknown')

numeric_columns = [
    'TOTAL_VACCINATIONS', 'PERSONS_VACCINATED_1PLUS_DOSE', 'TOTAL_VACCINATIONS_PER100',
    'PERSONS_VACCINATED_1PLUS_DOSE_PER100', 'PERSONS_LAST_DOSE', 'PERSONS_LAST_DOSE_PER100',
    'PERSONS_BOOSTER_ADD_DOSE', 'PERSONS_BOOSTER_ADD_DOSE_PER100'
]
vaccination_data[numeric_columns] = vaccination_data[numeric_columns].fillna(0)

vaccination_data['FIRST_VACCINE_DATE'] = vaccination_data['FIRST_VACCINE_DATE'].fillna('Unknown')

global_totals = vaccination_data[numeric_columns].sum()

plt.figure(figsize=(10, 6))
global_totals.drop(['TOTAL_VACCINATIONS_PER100', 'PERSONS_VACCINATED_1PLUS_DOSE_PER100',
                    'PERSONS_LAST_DOSE_PER100', 'PERSONS_BOOSTER_ADD_DOSE_PER100']).plot(
    kind='bar', color=['blue', 'green', 'red', 'purple'], alpha=0.7
)
plt.title('Global Vaccination Statistics', fontsize=16)
plt.ylabel('Count (in billions)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

regional_coverage = vaccination_data.groupby('WHO_REGION')[
    ['TOTAL_VACCINATIONS_PER100', 'PERSONS_VACCINATED_1PLUS_DOSE_PER100', 'PERSONS_BOOSTER_ADD_DOSE_PER100']
].mean()

regional_coverage.plot(kind='bar', figsize=(12, 6), alpha=0.8)
plt.title('Regional Vaccination Coverage (Per 100 People)', fontsize=16)
plt.ylabel('Vaccination Rate', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
