### Covid19 Data

```python
import json
import requests
import pandas as pd

API_KEY = "a60ce11ace3f42238d4cd82b098810a9"
BASE_URL = "https://api.covidactnow.org/v2/states.timeseries.json"

def fetch_json_data(api_key):
    try:
        response = requests.get(f"{BASE_URL}?apiKey={api_key}")
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

json_data = fetch_json_data(API_KEY)
```

```python
all_states_data = []

for state_data in json_data:
    state_name = state_data['state']
    daily_metrics = state_data['actualsTimeseries']
    
    state_df = pd.DataFrame(daily_metrics)
    state_df['date'] = pd.to_datetime(state_df['date'])  
    state_df['state'] = state_name  
    all_states_data.append(state_df)

df = pd.concat(all_states_data, ignore_index=True)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 6)
nationwide_totals = df.groupby('date')[['cases', 'deaths']].sum()

plt.plot(nationwide_totals.index, nationwide_totals['cases'], label='Total Cases', color='blue')
plt.plot(nationwide_totals.index, nationwide_totals['deaths'], label='Total Deaths', color='red')
plt.title("Nationwide COVID-19 Cases and Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.grid()
plt.show()
```

```python
selected_states = ['NY', 'CA', 'TX', 'FL']
state_data_filtered = df[df['state'].isin(selected_states)]
plt.figure(figsize=(12, 6))
for state in selected_states:
    state_df = state_data_filtered[state_data_filtered['state'] == state]
    plt.plot(state_df['date'], state_df['cases'], label=state)

plt.title("Daily COVID-19 Cases in Selected States")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid()
plt.show()
```

### Unemployment Data

```python
url = "https://api.stlouisfed.org/fred/series/observations"
api_key = "fdbf32479e0649ccea39dd7fcc5cf7ce"

params = {
    "series_id": "UNRATE",  
    "api_key": api_key,     
    "file_type": "json",    
    "observation_start": "2018-01-01",  
    "observation_end": "2024-12-31"     
}

response = requests.get(url, params=params)
if response.status_code == 200:
    data = response.json()
    if "observations" in data:
        unemployment_data = pd.DataFrame(data["observations"])
        unemployment_data['date'] = pd.to_datetime(unemployment_data['date'])
        unemployment_data['value'] = pd.to_numeric(unemployment_data['value'], errors='coerce')
        unemployment_data.rename(columns={"value": "unemployment_rate"}, inplace=True)
    
        print(unemployment_data.head())
        unemployment_data.to_csv("us_unemployment_rate.csv", index=False)
        print("Unemployment data saved to 'us_unemployment_rate.csv'")
    else:
        print("No observations data found in the response.")
else:
    print(f"Failed to fetch data. HTTP Status Code: {response.status_code}")
```
